*ANALISIS*

**Problema :** calcular la factorial de un numero

El factorial es el producto de los enteros no negrativos desde 1 hasta n:

n! = 1 x 2 x 3 x .... n
- Aplica para numeros enteros no negrativos -> Se debe validar

**Entrada(s)**:
- Un numero entero no negativo ingresado por el usuario.

**Salida(s)**:
- Un numero que represente el factorial del número ingresado por el usuario.
-variable de salida: f(de tipo floatante) -> El factorial de un número puede tomar un valor muy grande

**Otras variables:**
- El iterador utilizado en ciclo

**Proceso(s):**
- Se debe solicitar el numero
- Inicializar una variable donde se guardara el resultado
- Inicializar un contador para incrementar el producto en cada iteración.
- Iterar desde un valor inicial uno hasta llegar al valor dado por el usuario
- En cada iteracion, multiplicar y guardar el resultado
- Finalmente, se muestra el valor obtenido.

**Restricciones:**
- Aplica para numeros enteros no negrativos -> Se debe validar el signo
- la validacion se puede hacer con un condicional que verifique que el valor ingresado es un numero entero positivo. De lo contrario indicarle al usuario para que intente nuevamente o infalizar el programa. Como esto no se indica, se procedera a indicar el error y finalizar
- El factorial de 0 es 1.

***

*PSEUDOCODIGO*

    Program_calcular_factorial

    //Programa para calcular el factorial de un numero entero no negativo.
    //El factorial se calcula multiplicando los numero consecutivos desde el 1 hasta el valor n

    Start
    //Declaracion de variables
    Declare Integer n //Representa  el valor ingresado por el usuario
    Declare Real f //Representa el resultado/factorial
    Declare Integer i //representa el iterador


    Display "Programa para el calculo del factorial"
    Display "Ingrese un valor entero no negativo"
    Input n


    If n < 0 Then
      Display"El numero ingresado es un número negativo."
    Else
    //Defino el resultado por defecto
      Set f = 1

    For i = 1 to n
      f = f * i
    End For

    //se muestra el resultado
    Display "El factorial de",n,"es",f

    End If
    End

**Pregunta:**
"¿Cual es el factorial de 25?"

**Respuesta:** --------

```python
# Program_calcular_factorial

print("Programa para el calculo del factorial")
n = float(input("Ingrese un valor entero no negativo: "))
if n < 0:
	print("El numero ingresado es un número negativo.")
else:
  f = 1
for i in range(1.0, n + 1.0):
  f *= i
print("El factorial de", n, "es", f)
```
